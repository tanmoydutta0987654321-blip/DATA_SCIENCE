### Handling Missing Value

```python
data = {
    'Name':['Amit','Priya','Rohit',None,'Amit'],
    'Roll':[20,52,84,56,89],
    'Mark':[78,None,90,75,85]
}
df=pd.DataFrame(data)
print("Original dataframe :- \n",df)
# print(df.dtype)
```

#### Data Cleaning

```python
print("This is for isnull: \n",df.isnull()) 
print("This is for notnull: \n",df.notnull())
```

```python
print('Count the missing  values  per columns is : \n',df.isnull().sum())
print('Count total missing  values  in df is :',df.isnull().sum().sum())
```

##### Filling Missing Values

```python
print(df.fillna('Unknown'))
df['Mark']=df['Mark'].fillna(56)
# df['Marks']=df['Marks'].fillna(df['Mark'].mean())
print(df)
```
