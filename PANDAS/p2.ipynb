# Indexing & Slicing

```python
s=pd.Series(10,index=['a','b','c','d','e'])
print(s)
```

```python
s=pd.Series([45,559,54541,84512],index=['a',89,'tanmoy',4+7j])
print(s)
```

```python
y=pd.Series([7+8j,'95',97.52,84,546,78], index=['T','A','N','M','O','Y'])
y.name='marks'
print(y)
print(y['T'])
print(y.loc['A':'O'])
```

```python
print(y.iloc[0])
print(y.iloc[0:3])
print(y.iloc[:4])
```

```python
Ages=pd.Series([25,46,78,69],index=['ram','gun','kol','jonny'])
old=Ages[Ages>28]
print(old)
```
